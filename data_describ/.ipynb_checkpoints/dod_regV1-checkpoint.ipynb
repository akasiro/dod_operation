```python

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.gridspec as gridspec
workspace = r'/home/ubuntu/dod_aws/work'
data_path = os.path.join(workspace,r'datasets/derived/reg20200206')
lv1_mau_csv = os.path.join(data_path,'dod_lv1_mau20200206.csv')
lv1_cover_csv = os.path.join(data_path, 'dod_lv1_cover20200206.csv')
lv2_csv = os.path.join(data_path, 'dod_lv2_catedemand20200206.csv')
```

# 1.read data

```python
lv1_mau = pd.read_csv(lv1_mau_csv)
lv1_cover = pd.read_csv(lv1_cover_csv)
lv2 = pd.read_csv(lv2_csv)
```

# 2. pre analysis

```python
plt.figure()
sns.kdeplot(lv1_mau['mau'], legend=False)
```

```python

f = plt.figure()
type_code_set = list(set(lv1_mau['type_code'].values.tolist()))


f,axes = plt.subplots(1,2, sharey = True)
for i in range(2):
    
    tmp = lv1_mau[lv1_mau['type_code'] == type_code_set[i]]
    sns.kdeplot(tmp['mau'], ax = axes[i])
    axes[i].set_title(type_code_set[i])

plt.tight_layout()
plt.show()
```
